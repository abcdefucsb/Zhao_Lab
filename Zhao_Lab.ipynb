```python
import pandas as pd

# Load the Excel file
file_path = 'oncogene.xlsx'  # replace with your file path
df = pd.read_excel(file_path)

# Extract the data from a specific column (replace 'ColumnName' with your actual column name)
column_data = df['GRCh38 RefSeq'].dropna()

# Display the column data
print(column_data)
#-------------------------------------------------------------------------------------------------------------------------------------------
from Bio import Entrez

# Function to fetch FASTA file from NCBI
def fetch_fasta(gene_id, email):
    Entrez.email = email  # Always provide an email to NCBI API
    try:
        # Fetch sequence from NCBI
        handle = Entrez.efetch(db="nucleotide", id=gene_id, rettype="fasta", retmode="text")
        fasta_data = handle.read()
        handle.close()
        return fasta_data
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example: Fetching a sequence for human BRCA1 gene
gene_ids = column_data.tolist()  # Example Gene ID for BRCA1 (can be any NCBI gene ID)
fasta_data = []
processed_genes = []
email="shiyang.xin@ucsf.edu"


# Open the file to write the remaining gene IDs
with open("remaining_gene_ids.txt", "w") as remaining_file:
    for gene_id in gene_ids:
        fasta_sequence = fetch_fasta(gene_id, email)
        if fasta_sequence:
            try:
                # Save FASTA sequence to a file named after the gene ID
                with open(f"{gene_id}.fasta", "w") as file:
                    file.write(fasta_sequence)
                print(f"FASTA for {gene_id} saved successfully.")
                processed_genes.append(gene_id)
            except Exception as file_error:
                print(f"Failed to save {gene_id}: {file_error}")
        else:
            print(f"No FASTA data found for {gene_id}.")

        # After processing each gene, remove it from gene_ids and update the file
        remaining_genes = [gid for gid in gene_ids if gid not in processed_genes]
        remaining_file.seek(0)  # Go to the start of the file to overwrite
        remaining_file.truncate(0)  # Clear the file content
        remaining_file.writelines("\n".join(remaining_genes))


print("Script completed. Remaining gene IDs saved to 'remaining_gene_ids.txt'.")
```

```python
import os
import zipfile
from google.colab import files

# Create a zip file for all .fasta files
output_zip = "fasta_files.zip"

# Check for existing FASTA files
fasta_files = [f for f in os.listdir() if f.endswith(".fasta")]

if fasta_files:
    with zipfile.ZipFile(output_zip, "w") as zipf:
        for fasta_file in fasta_files:
            zipf.write(fasta_file)
    print(f"All FASTA files have been zipped into {output_zip}.")

    # Provide download link
    files.download(output_zip)
else:
    print("No FASTA files found to zip.")

```
